## Data Cleaning

#### Handling Missing Data

```python
import numpy as np
import pandas as pd
```

```python
df =  pd.read_csv('NFL Play by Play.csv')
```

```python
df.head()
```

```python
missing_values_count = df.isnull().sum()
missing_values_count[0:10]
```

```python
#total percentage of data
total_cells = np.product(df.shape)

total_missing = missing_values_count.sum()

percentage_missing = (total_missing/total_cells)*100
print(percentage_missing)
```

```python
#drop all rows tha have missing data
df.dropna()
```

```python
#drop all columns that have missing data
column_with_ns_dropped = df.dropna(axis=1)
column_with_ns_dropped.head()
```

```python
#how much data was deleted?
print('columns in original data : %d\n' %df.shape[1])
print('columns in new data : %d\n' %column_with_ns_dropped.shape[1])
```

```python
#fill data at null place
subset__df = df.loc[:,'EPA':'Season'].head()
subset__df
```

```python
subset__df.fillna(0)
```

```python
#fill just next value to the back null value
subset__df.fillna(method='bfill', axis=0).fillna(0)
```

```python
# fill with mean value
mean_airwpa = subset__df['airWPA'].mean()
print(mean_airwpa)
subset__df.airWPA.fillna(mean_airwpa)
```

```python
# fill with median value
median_airwpa = subset__df['airWPA'].median()
print(median_airwpa)
subset__df.airWPA.fillna(median_airwpa)
```
